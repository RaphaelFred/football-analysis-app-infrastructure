```python
import pandas as pd
```

```python
df_ec2 = pd.read_csv('/Users/raphael/Downloads/over25_results_2023-03-24-ec2.csv')
df_local = pd.read_csv('/Users/raphael/Downloads/over25_results_2023-03-24-local.csv')
```

```python
ec2_fixtures = set(df_ec2['fixture_id'])
local_fixtures = set(df_local['fixture_id'])
```

```python
fixtures_not_ec2 = local_fixtures - ec2_fixtures
pd.DataFrame.to_csv(df_local[df_local['fixture_id'].isin(fixtures_not_ec2)], "not_ec2.csv")
```
